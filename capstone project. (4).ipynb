```python
from bs4 import BeautifulSoup
import requests
link='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
table= requests.get(link).text
```

```python
soup = BeautifulSoup(table, 'html.parser')
```

```python
tab = []
for tr in soup.tbody.find_all('tr'):
    tab.append([ td.get_text().strip() for td in tr.find_all('td')])
```

```python
df = pd.DataFrame(tab, columns=['PostalCode','Borough','Neighborhood'])
df.head()
```

```python
df = df.dropna()
```

```python
NA = 'Not assigned'
df = df[(df.PostalCode != NA ) & (df.Borough != NA) & (df.Neighborhood != NA)]
df.head()
```

```python
def NHlist(groups):    
    return ', '.join(sorted(groups['Neighborhood'].tolist()))
                    
gp = df.groupby(['PostalCode', 'Borough'])
dataframe = gp.apply(NHlist).reset_index(name='Neighborhood')

dataframe.head()
```

```python

!pip install geocoder
import geocoder
def latlng(postal_code):
    coords = None
    while(coords is None):
        maps = geocoder.arcgis('{}, Toronto, Ontario'.format(postal_code))
        coords = maps.latlng
    return coords
```

```python
postalCodes = df['PostalCode']    
cords = [ latlng(postal_code) for postal_code in postalCodes.tolist() ]
```

```python
dataframe_coords = pd.DataFrame(cords, columns=['Latitude', 'Longitude'])
dataframe['Latitude'] = dataframe_coords['Latitude']
dataframe['Longitude'] = dataframe_coords['Longitude']
```
